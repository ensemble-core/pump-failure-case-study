```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import xgboost as xgb
from sklearn.metrics import classification_report, roc_auc_score, roc_curve, auc, f1_score
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
import torch
import torch.nn as nn
import torch.optim as optim
from torch.nn import functional as F
from torch import Tensor
import random
import traceback
import math
import time
from itertools import product
import statistics
import shap
import copy
from collections import Counter # For argmax plot counts

# --- Configuration ---
DATA_PATH = '/Users/xianghe/Downloads/pump_failure_dataset.csv'
SEED = 42
TEST_SIZE = 0.3
# NN Training Params
NDLINEAR_EPOCHS = 100 
NDLINEAR_LR = 0.001
# XGBoost Params
XGB_PARAMS_FIXED = {'n_estimators': 100, 'max_depth': 3, 'learning_rate': 0.1, 'eval_metric':'logloss', 'use_label_encoder':False, 'random_state': SEED}

# --- Grid Search Parameter Grid ---
param_grid = {
    'nd_core_hidden_size': [(8, 1), (16, 1), (24, 1), (32, 1), (40, 1), (48, 1)],
    'nd_core_feature_output_dim': [8, 16, 24, 32, 40, 48], 
    'dropout_rate': [0.1, 0.15, 0.2, 0.25, 0.3, 0.4]
}

EVALUATION_METRIC = 'f1_score' 

# --- Set seed ---
random.seed(SEED); np.random.seed(SEED); torch.manual_seed(SEED)
if torch.cuda.is_available(): torch.cuda.manual_seed_all(SEED); device = torch.device("cuda:0"); print(f"Using GPU: {device}")
else: device = torch.device("cpu"); print("Using CPU.")
torch.set_default_dtype(torch.float32)

# --- 1. Define NdLinear Class Directly ---
class NdLinear(nn.Module):
    def __init__(self, input_dims: tuple, hidden_size: tuple, transform_outer=True):
        super(NdLinear, self).__init__();
        if len(input_dims) != len(hidden_size): raise ValueError("Input/hidden shape length mismatch.")
        self.input_dims=input_dims; self.hidden_size=hidden_size; self.num_layers=len(input_dims); self.transform_outer=transform_outer
        self.align_layers = nn.ModuleList([nn.Linear(input_dims[i], hidden_size[i]) for i in range(self.num_layers)])
    def forward(self, X):
        if len(X.shape) != self.num_layers + 1: raise ValueError(f"Input rank mismatch...")
        for i in range(self.num_layers):
             if self.input_dims[i] != -1 and X.shape[i+1] != self.input_dims[i]: raise ValueError(f"Input dim mismatch...")
        X_intermediate = X; num_transforms = self.num_layers
        for i in range(num_transforms):
            if self.transform_outer: layer_idx, transpose_dim = i, i + 1
            else: layer_idx, transpose_dim = num_transforms - (i + 1), num_transforms - i
            layer = self.align_layers[layer_idx]; target_hidden_dim = self.hidden_size[layer_idx]
            if self.input_dims[layer_idx] == -1: continue
            X_intermediate = torch.transpose(X_intermediate, transpose_dim, self.num_layers).contiguous()
            original_shape_prefix = X_intermediate.shape[:-1]; current_last_dim_size = X_intermediate.shape[-1]
            X_flat = X_intermediate.view(-1, current_last_dim_size); X_transformed = layer(X_flat)
            X_intermediate = X_transformed.view(*original_shape_prefix, target_hidden_dim)
            X_intermediate = torch.transpose(X_intermediate, transpose_dim, self.num_layers).contiguous()
        return X_intermediate

# --- 2. Define PyTorch Feature Extractor ---
class NdFeatureExtractor(nn.Module):
    # --- (Keep definition adapted from user's ndMLP) ---
     def __init__(self, input_shape, hidden_shape, embedding_shape, dropout_rate=0.2):
        super().__init__()
        self.nd1 = NdLinear(input_dims=tuple(input_shape), hidden_size=tuple(hidden_shape))
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(dropout_rate)
        self.nd2 = NdLinear(input_dims=tuple(hidden_shape), hidden_size=tuple(embedding_shape))
     def forward(self, x):
        x_nd1_out = self.nd1(x); x_relu = self.relu(x_nd1_out)
        x_nd2_out = self.nd2(x_relu)
        x_flat = x_nd2_out.view(x_nd2_out.size(0), -1)
        features = self.dropout(x_flat); return features

class NdClassifier(nn.Module):
    def __init__(self, feature_extractor: NdFeatureExtractor, num_classes=2):
        super(NdClassifier, self).__init__(); self.feature_extractor = feature_extractor; feature_output_dim = -1
        # Calculate expected flat output dim from feature extractor's second NdLinear layer
        if hasattr(feature_extractor, 'nd2') and hasattr(feature_extractor.nd2, 'hidden_size'):
             feature_output_dim = int(np.prod(feature_extractor.nd2.hidden_size))
        else: raise ValueError("Could not determine feature_output_dim from extractor.")
        self.classifier = nn.Linear(feature_output_dim, num_classes)
    def forward(self, x): features = self.feature_extractor(x); logits = self.classifier(features); return logits
    def extract_features(self, x):
        self.feature_extractor.eval();
        with torch.no_grad(): features = self.feature_extractor(x); return features

# --- 3. Define Feature Lineage Function ---
def compute_feature_lineage(model: NdFeatureExtractor, input_tensor: Tensor):
    model.eval(); input_tensor_clone = input_tensor.clone().detach().requires_grad_(True)
    output_features_flat = model(input_tensor_clone); B = input_tensor_clone.shape[0]
    O = output_features_flat.shape[1]; input_flat_dim = input_tensor_clone.shape[1] * input_tensor_clone.shape[2]
    lineage = torch.zeros(B, O, input_flat_dim, device=input_tensor_clone.device, dtype=input_tensor_clone.dtype)
    print(f"  Calculating gradients for {O} output features...")
    for i in range(O):
        if input_tensor_clone.grad is not None: input_tensor_clone.grad.zero_()
        grads = torch.autograd.grad(outputs=output_features_flat[:, i], inputs=input_tensor_clone, grad_outputs=torch.ones_like(output_features_flat[:, i]), retain_graph=True, create_graph=False, allow_unused=True)[0]
        if grads is not None: grads_flat = grads.reshape(B, -1); lineage[:, i, :] = grads_flat.abs()
    return lineage

# --- 4. Data Loading and Feature Engineering ---
print("Loading and preprocessing data...");
try: df = pd.read_csv(DATA_PATH)
except FileNotFoundError: print(f"Error: Data file not found at {DATA_PATH}"); exit()
df['timestamp'] = pd.to_datetime(df['timestamp'], format='%Y-%m-%d %H:%M:%S')
core_features = ['vibration_level', 'temperature_C', 'pressure_PSI', 'flow_rate_m3h']
engineered_feature_names = []
window_size_eng = 5
for feature in core_features:
    roll_col=f'{feature}_rolling_mean_{window_size_eng}'; trend_col=f'{feature}_trend_{window_size_eng}'
    df[roll_col] = df[feature].rolling(window=window_size_eng).mean(); df[trend_col] = df[feature] - df[feature].shift(window_size_eng)
    engineered_feature_names.extend([roll_col, trend_col])
df['hour'] = df['timestamp'].dt.hour; df['day_of_week'] = df['timestamp'].dt.dayofweek
engineered_feature_names.extend(['hour', 'day_of_week'])
df = df.dropna().reset_index(drop=True)
y = df['failure']; X = df.drop(columns=['timestamp', 'failure'])
all_feature_names = list(X.columns); num_all_features = X.shape[1]
print(f"Total features: {num_all_features}")
core_feature_indices = [all_feature_names.index(f) for f in core_features]

# --- 5. Data Splitting ---
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=TEST_SIZE, random_state=SEED, stratify=y)
print(f"Train samples: {len(X_train)}, Test samples: {len(X_test)}")

# --- 6. Scaling ---
scaler = StandardScaler(); X_train_scaled = scaler.fit_transform(X_train); X_test_scaled = scaler.transform(X_test)
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=all_feature_names, index=X_train.index)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=all_feature_names, index=X_test.index)

# --- 7. Train and Evaluate Baseline XGBoost ---
print("\n--- Training & Evaluating Baseline XGBoost ---")
smote_baseline = SMOTE(random_state=SEED); X_train_resampled_baseline, y_train_resampled_baseline = smote_baseline.fit_resample(X_train_scaled, y_train)
xgb_baseline = xgb.XGBClassifier(**XGB_PARAMS_FIXED); xgb_baseline.fit(X_train_resampled_baseline, y_train_resampled_baseline)
xgb_baseline_preds = xgb_baseline.predict(X_test_scaled); xgb_baseline_proba = xgb_baseline.predict_proba(X_test_scaled)[:, 1]
baseline_roc_auc = roc_auc_score(y_test, xgb_baseline_proba); baseline_f1 = f1_score(y_test, xgb_baseline_preds, pos_label=1)
print("Baseline XGBoost Classification Report:"); print(classification_report(y_test, xgb_baseline_preds))
print(f"Baseline XGBoost ROC AUC: {baseline_roc_auc:.4f}"); print(f"Baseline XGBoost F1 Score (Class 1): {baseline_f1:.4f}")

# --- 8. Grid Search Loop ---
print("\n--- Starting Grid Search for NdLinear (Core) Hybrid Model ---")
grid_search_results = []
search_start_time = time.time()
param_combinations = list(product(param_grid['nd_core_hidden_size'], param_grid['nd_core_feature_output_dim'], param_grid['dropout_rate']))
best_score = -1.0; best_params = None; best_nd_feature_extractor = None # Store best extractor
best_hybrid_xgb_model = None; best_X_test_combined = None; best_X_test_nd_core_tensor = None

# --- Prepare unchanging data parts ---
X_train_core_scaled = X_train_scaled_df[core_features].values; X_test_core_scaled = X_test_scaled_df[core_features].values
core_input_shape = (len(core_features), 1)
X_train_nd_input_core = X_train_core_scaled.reshape(-1, *core_input_shape); X_test_nd_input_core = X_test_core_scaled.reshape(-1, *core_input_shape)
X_train_full_nd_core_tensor = torch.tensor(X_train_nd_input_core, dtype=torch.float32).to(device); X_test_nd_core_tensor = torch.tensor(X_test_nd_input_core, dtype=torch.float32).to(device)
X_train_engineered_scaled = X_train_scaled_df[engineered_feature_names].values; X_test_engineered_scaled = X_test_scaled_df[engineered_feature_names].values
n_samples_train, d1_core, d2_core = X_train_nd_input_core.shape
X_train_nd_flat_core = X_train_nd_input_core.reshape(n_samples_train, -1)
smote_nd_core = SMOTE(random_state=SEED); X_train_resampled_nd_flat_core, y_train_resampled_nd_core = smote_nd_core.fit_resample(X_train_nd_flat_core, y_train)
X_train_resampled_nd_core_tensor = torch.tensor(X_train_resampled_nd_flat_core.reshape(-1, d1_core, d2_core), dtype=torch.float32).to(device)
y_train_resampled_nd_core_tensor = torch.tensor(y_train_resampled_nd_core.values, dtype=torch.long).to(device)
# --- End unchanging data prep ---

for i, (hs, fod, dr) in enumerate(param_combinations):
    current_params = {'hidden_size': hs, 'feature_output_dim': fod, 'dropout_rate': dr}
    print(f"\n[{i+1}/{len(param_combinations)}] Testing Params: {current_params}")
    iteration_start_time = time.time()
    try:
        # 1. Instantiate NdFeatureExtractor/NdClassifier
        current_embedding_shape = (fod, 1) # Assume final feature dim maps to this shape for nd2
        current_hidden_shape = hs
        if int(np.prod(current_hidden_shape)) == 0 or int(np.prod(current_embedding_shape)) == 0: continue # Skip invalid

        feature_extractor_gs = NdFeatureExtractor(input_shape=core_input_shape, hidden_shape=current_hidden_shape, embedding_shape=current_embedding_shape, dropout_rate=dr).to(device)
        nd_classifier_gs = NdClassifier(feature_extractor_gs, num_classes=2).to(device)

        # 2. Train NdClassifier
        criterion_gs = nn.CrossEntropyLoss(); optimizer_gs = optim.Adam(nd_classifier_gs.parameters(), lr=NDLINEAR_LR)
        for epoch in range(NDLINEAR_EPOCHS): # Use NDLINEAR_EPOCHS here
            nd_classifier_gs.train(); optimizer_gs.zero_grad(); outputs = nd_classifier_gs(X_train_resampled_nd_core_tensor)
            loss = criterion_gs(outputs, y_train_resampled_nd_core_tensor); loss.backward(); optimizer_gs.step()

        # 3. Extract Features
        train_feats_nd_core_gs = nd_classifier_gs.extract_features(X_train_full_nd_core_tensor).cpu().numpy()
        test_feats_nd_core_gs = nd_classifier_gs.extract_features(X_test_nd_core_tensor).cpu().numpy()

        # 4. Combine Features
        X_train_combined_gs = np.concatenate((X_train_engineered_scaled, train_feats_nd_core_gs), axis=1)
        X_test_combined_gs = np.concatenate((X_test_engineered_scaled, test_feats_nd_core_gs), axis=1)

        # 5. SMOTE Combined Train
        smote_combined_gs = SMOTE(random_state=SEED); X_train_resampled_combined_gs, y_train_resampled_combined_gs = smote_combined_gs.fit_resample(X_train_combined_gs, y_train)

        # 6. Train XGBoost
        xgb_on_combined_gs = xgb.XGBClassifier(**XGB_PARAMS_FIXED); xgb_on_combined_gs.fit(X_train_resampled_combined_gs, y_train_resampled_combined_gs)

        # 7. Evaluate XGBoost
        xgb_combined_preds_gs = xgb_on_combined_gs.predict(X_test_combined_gs); xgb_combined_proba_gs = xgb_on_combined_gs.predict_proba(X_test_combined_gs)[:, 1]
        roc_auc = roc_auc_score(y_test, xgb_combined_proba_gs); f1 = f1_score(y_test, xgb_combined_preds_gs, pos_label=1)

        # 8. Store Result AND Check if it's the best so far
        current_score = f1 if EVALUATION_METRIC == 'f1_score' else roc_auc
        results_entry = {'params': current_params, 'roc_auc': roc_auc, 'f1_score': f1, 'eval_score': current_score}
        grid_search_results.append(results_entry); print(f"    Result: ROC AUC={roc_auc:.4f}, F1={f1:.4f} (took {time.time() - iteration_start_time:.1f}s)")

        if current_score > best_score:
            print(f"    >>>> New best {EVALUATION_METRIC} found: {current_score:.4f} <<<<")
            best_score = current_score; best_params = current_params
            # Save the feature extractor INSTANCE from the best run
            best_nd_feature_extractor = copy.deepcopy(feature_extractor_gs)
            best_hybrid_xgb_model = xgb_on_combined_gs
            best_X_test_combined = X_test_combined_gs
            best_X_test_nd_core_tensor = X_test_nd_core_tensor # Save core features tensor for lineage

    except Exception as e: print(f"    ERROR for {current_params}: {e}"); print(traceback.format_exc()); grid_search_results.append({'params': current_params, 'roc_auc': 0.0, 'f1_score': 0.0, 'eval_score': 0.0, 'error': str(e)})

# --- 9. Analyze Grid Search Results ---
search_duration = time.time() - search_start_time
print(f"\n--- Grid Search Completed in {search_duration:.2f} seconds ---")
if not grid_search_results: print("No results!"); exit()
results_df = pd.DataFrame(grid_search_results); params_df = pd.json_normalize(results_df['params'])
results_df = pd.concat([params_df.reset_index(drop=True), results_df.drop('params', axis=1)], axis=1)
results_df = results_df.sort_values(by='eval_score', ascending=False).reset_index(drop=True)
print("\n--- Grid Search Results (Top 5) ---"); print(results_df.head())
if best_params is None: print("\nNo best model saved!"); exit()
print("\n--- Best Parameters Found ---"); print(f"Based on Metric: {EVALUATION_METRIC}"); print(best_params)

# --- 10. Explainability Analysis ---
print("\n--- Performing Explainability Analysis ---")

# ** SHAP for Baseline XGBoost **
print("  Calculating SHAP values for Baseline XGBoost...")
try:
    X_test_scaled_df_shap = pd.DataFrame(X_test_scaled, columns=all_feature_names)
    explainer_base = shap.Explainer(xgb_baseline, X_train_resampled_baseline); shap_values_base = explainer_base(X_test_scaled)
    print("  Generating SHAP summary plot (bar) for Baseline XGBoost..."); plt.figure(); shap.summary_plot(shap_values_base.values, X_test_scaled_df_shap, plot_type="bar", show=False); plt.title("SHAP Feature Importance (Baseline XGBoost)"); plt.tight_layout(); plt.show()
    print("  Generating SHAP summary plot (beeswarm) for Baseline XGBoost..."); plt.figure(); shap.summary_plot(shap_values_base.values, X_test_scaled_df_shap, show=False); plt.title("SHAP Feature Summary (Baseline XGBoost)"); plt.tight_layout(); plt.show()
except Exception as e: print(f"  Error during Baseline SHAP analysis: {e}")

# --- 11. Explainability Analysis ---
print("\n--- Performing Explainability Analysis ---")

# ** SHAP for Baseline XGBoost **
print("  Calculating SHAP values for Baseline XGBoost...")
try:
    X_test_scaled_df_shap = pd.DataFrame(X_test_scaled, columns=all_feature_names)
    explainer_base = shap.Explainer(xgb_baseline, X_train_resampled_baseline); shap_values_base = explainer_base(X_test_scaled)
    print("  Generating SHAP summary plot (bar) for Baseline XGBoost..."); plt.figure(); shap.summary_plot(shap_values_base.values, X_test_scaled_df_shap, plot_type="bar", show=False); plt.title("SHAP Feature Importance (Baseline XGBoost)"); plt.tight_layout(); plt.show()
except Exception as e: print(f"  Error during Baseline SHAP analysis: {e}")


# ** Feature Lineage for Best NdLinear Extractor **
print("\n  Calculating Feature Lineage for Best NdLinear Extractor...")
lineage_tensor = None # Initialize
if best_nd_feature_extractor and best_X_test_nd_core_tensor is not None:
    try:
        lineage_tensor = compute_feature_lineage(best_nd_feature_extractor, best_X_test_nd_core_tensor)
        print(f"  Computed Lineage Tensor shape: {lineage_tensor.shape}")
        output_dim_O = lineage_tensor.shape[1]; input_dim_I = lineage_tensor.shape[2]

        # --- Plot 1: Aggregate Importance ---
        aggregate_importance = lineage_tensor.abs().sum(dim=1).mean(dim=0).cpu().numpy()
        if len(aggregate_importance) == len(core_features):
            print("  Plotting: Feature Lineage Aggregate Importance...")
            plt.figure(figsize=(8, 4)); plt.bar(core_features, aggregate_importance); plt.xlabel("Core Input Features"); plt.ylabel("Mean Abs Gradient (Contribution)"); plt.title("NdLinear Lineage: Overall Input Importance"); plt.xticks(rotation=15); plt.tight_layout(); plt.show()

        # --- Calculate Mean Abs Lineage (used in multiple plots) ---
        mean_abs_lineage_np = lineage_tensor.abs().mean(dim=0).cpu().numpy() # Shape (O, I=4)

        # --- Plot 2: Heatmap of Mean Absolute Gradients ---
        if mean_abs_lineage_np is not None: # Changed check from lineage_tensor
            print("  Plotting: Feature Lineage Heatmap (Input Features on Y-axis)...")
            data_to_plot = mean_abs_lineage_np.T
            plt.figure(figsize=(max(8, output_dim_O / 4), 6)); plt.imshow(data_to_plot, cmap='viridis', aspect='auto'); plt.colorbar(label='Mean Absolute Gradient')
            plt.ylabel("Core Input Feature"); plt.xlabel("Extracted Feature Index"); plt.yticks(ticks=np.arange(input_dim_I), labels=core_features); plt.xticks(ticks=np.arange(0, output_dim_O, max(1, output_dim_O // 10))); plt.title("NdLinear Lineage: Input Influence on Output Features"); plt.tight_layout(); plt.show()

        # --- Plot 3: Input Contributions for Specific Output Features ---
        if mean_abs_lineage_np is not None:
            print("  Plotting: Input Contributions for Example Output Features...")
            num_outputs_to_plot = min(5, output_dim_O); plt.figure(figsize=(12, 3 * num_outputs_to_plot))
            for i in range(num_outputs_to_plot):
                plt.subplot(num_outputs_to_plot, 1, i + 1); output_feature_gradients = mean_abs_lineage_np[i, :]; plt.bar(core_features, output_feature_gradients); plt.ylabel("Mean Abs Gradient"); plt.title(f"Input Contributions to Output Feature ND_{i}"); plt.xticks(rotation=15)
            plt.tight_layout(); plt.show()

        # --- Plot 4: Output Sensitivity for Specific Input Features ---
        if mean_abs_lineage_np is not None:
             print("  Plotting: Output Sensitivity for Example Input Features...")
             num_inputs_to_plot = min(4, input_dim_I); plt.figure(figsize=(12, 3 * num_inputs_to_plot))
             for i in range(num_inputs_to_plot):
                  plt.subplot(num_inputs_to_plot, 1, i + 1); input_feature_influence = mean_abs_lineage_np[:, i]; plt.plot(range(output_dim_O), input_feature_influence); plt.xlabel("Output Feature Index"); plt.ylabel("Mean Abs Gradient"); plt.title(f"Output Sensitivity to Input Feature '{core_features[i]}'"); plt.grid(axis='y', linestyle='--', alpha=0.7)
             plt.tight_layout(); plt.show()

        # --- Plot 5: Input Contributions for a SINGLE Output Feature ---
        if mean_abs_lineage_np is not None:
            target_output_index = 47 # Example - adjust if needed
            print(f"\n  Plotting: Input Contributions for Specific Output Feature ND_{target_output_index}...")
            if 0 <= target_output_index < output_dim_O:
                single_output_gradients = mean_abs_lineage_np[target_output_index, :]
                plt.figure(figsize=(8, 4)); plt.barh(core_features, single_output_gradients); plt.xlabel("Mean Abs Gradient (Contribution)"); plt.ylabel("Core Input Feature"); plt.title(f"Input Contributions to Output Feature ND_{target_output_index}"); plt.tight_layout(); plt.show()
            else: print(f"  Warning: Output feature index {target_output_index} is out of bounds (0-{output_dim_O-1}).")

    except Exception as e: print(f"  Error during NdLinear Feature Lineage analysis: {e}"); traceback.print_exc()
else: print("  Could not perform Feature Lineage analysis (best extractor or data not available).")


# ** SHAP for BEST Hybrid XGBoost **
print("\n  Calculating SHAP values for Best Hybrid XGBoost...")
if best_hybrid_xgb_model and best_X_test_combined is not None:
    try:
        best_fod_shap = best_params['feature_output_dim']; nd_feature_names_final = [f"ND_{i}" for i in range(best_fod_shap)]
        combined_feature_names_final = engineered_feature_names + nd_feature_names_final
        X_test_combined_final_df_shap = pd.DataFrame(best_X_test_combined, columns=combined_feature_names_final)
        best_nd_train_feats = best_nd_feature_extractor.extract_features(X_train_full_nd_core_tensor).cpu().numpy()
        best_X_train_combined = np.concatenate((X_train_engineered_scaled, best_nd_train_feats), axis=1)
        smote_for_shap = SMOTE(random_state=SEED)
        X_train_resampled_shap_bg, _ = smote_for_shap.fit_resample(best_X_train_combined, y_train)
        background_data = shap.sample(X_train_resampled_shap_bg, 100) # Use 100 samples for background
        explainer_hybrid = shap.Explainer(best_hybrid_xgb_model, background_data)
        shap_values_hybrid = explainer_hybrid(X_test_combined_final)

        print("  Generating SHAP summary plot (bar) for Best Hybrid XGBoost..."); plt.figure(); shap.summary_plot(shap_values_hybrid.values, X_test_combined_final_df_shap, plot_type="bar", show=False); plt.title(f"SHAP Feature Importance (Best Hybrid XGBoost)"); plt.tight_layout(); plt.show()
        print("  Generating SHAP summary plot (beeswarm) for Best Hybrid XGBoost..."); plt.figure(); shap.summary_plot(shap_values_hybrid.values, X_test_combined_final_df_shap, show=False); plt.title(f"SHAP Feature Summary (Best Hybrid XGBoost)"); plt.tight_layout(); plt.show()
    except Exception as e: print(f"  Error during Hybrid SHAP analysis: {e}")
else: print("  Could not perform SHAP analysis on best hybrid model.")


# --- 12. Final Comparison ---
print("\n--- Final Model Comparison (Metrics) ---")
print(f"Baseline XGBoost ROC AUC: {baseline_roc_auc:.4f}"); print(f"Baseline XGBoost F1 Score: {baseline_f1:.4f}")
print("-" * 40)
best_hybrid_roc = results_df.loc[0, 'roc_auc'] if not results_df.empty and 'roc_auc' in results_df.columns else 'N/A'
best_hybrid_f1 = results_df.loc[0, 'f1_score'] if not results_df.empty and 'f1_score' in results_df.columns else 'N/A'
print(f"Best Hybrid Model ROC AUC: {best_hybrid_roc:.4f}")
print(f"Best Hybrid Model F1 Score: {best_hybrid_f1:.4f}")
if best_params: print(f"(Using Params: {best_params})")
print("-" * 40)
print("\nAnalysis complete.")
```
